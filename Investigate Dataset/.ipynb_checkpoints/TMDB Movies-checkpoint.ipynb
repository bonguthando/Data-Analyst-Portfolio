# Investigating a Movie Data Set

### Introduction

I am working on a movie data set collected from TMDb containing 10,000 movies with attributes such as movie budget, revenue, runtime and much more.
**NOTE:** This data analysis will focus on the top 20 movies with highest budget.

**QUESTIONS:**

1. How much did the top 20 movies cost to make?
2. What genres are represented in the top 20?
3. What is the average runtime of the top 20 movies.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#First we'll read and display a few lines from the csv file.
df = pd.read_csv('tmdb_movies.csv')
df.head()
```

```python
#Showing the description of the dataset
df.describe()
```

```python
#Getting information
df.info()
```

```python
#After displaying information of the complete data set, I would like to keep only
#the columns that would be needed to answer the questions above by dropping the rest of the columns.

df2 = df.drop(columns = ['id','imdb_id','popularity','cast','homepage','tagline','keywords','overview','vote_count','budget_adj','revenue_adj'])
```

```python
#Now that the unwanted columns have been dropped, I will display the top 20 movies based on revenue.

df2.nlargest(n=20, columns=['revenue'], keep='all')
```

```python
#'data' will be the focus of this analysis which is the top 20 movies with the highest revenue.
data = df2.nlargest(n=20, columns=['revenue'], keep='all')
```

```python
data.head(20)
```

```python
#check to see if there are any missing values.
data.isnull()
```

```python
#check for the total number of duplicates if there are any in the data I'll be working with.
sum(data.duplicated())
```

```python
#verifying to see if indeed we will be only working on the top 20 movies based on revenue.
data.info()
```

```python
data.describe()
```

## EDA (Exploratory Data Analysis)

> **QUESTION 1: How much did the top twenty movies cost to make?**

```python
#only Display the budget and name of each movie. (from highest budget to lowest budget.)
data[['budget','original_title']]
data_order = data[['budget','original_title']]
data_order.sort_values(by=['budget'],ascending = False)
```

> **QUESTION 2: How much profit did each movie make?**

```python
#Calculate the profit of each movie.
data['profit_made'] = data['revenue'] - data['budget']
#Add and display "Profit_made" column to the table.
data[['budget','revenue','profit_made','original_title']]

#display from highest profit to lowest profit
profit_data = data.sort_values(by='profit_made', ascending = False) [['budget','revenue','profit_made','original_title','release_year']]
profit_data.head(20)
```

```python
#plotting a histogram of movies vs revenue

#Size (width and height)of the figure
plt.figure(figsize=(15,8))
plt.bar( "original_title","revenue", data = data, color = 'red')
plt.title('twenty movies by revenue')
plt.xlabel('Movies', fontsize = 15)
plt.xticks(rotation = 90)
plt.ylabel('revenue', fontsize = 15)


plt.show()
 
```

The heighest earning movies here have made at 1,000,000,000.00USD. A majority of the movies were released after the year 2010.

```python
#plotting a histogram of movies vs profit 

#Size (width and height)of the figure
plt.figure(figsize=(15,8))
plt.bar( "original_title","profit_made", data = data, color = 'red')
plt.title('twenty movies by profit')
plt.xlabel('Movies', fontsize = 15)
plt.xticks(rotation = 90)
plt.ylabel('profit_made', fontsize = 15)


plt.show()
```

The movies by profit graph shows that while movies like 'Avatar', 'Star Wars: The Force Awakens', and 'Titanic' remain in their positions in the top twenty when comparing revenue to profit. 'The Avengers' and 'Iron Man 3' show lower profits when comapred to 'Jurassic World' and 'Minions' respectively.

> **QUESTION 3a: What is the runtime of the top twenty movies**

```python
#only Display the runtime and name of each movie. (from highest runtime to lowest runtime.)
data[['runtime','original_title']]
runtime_order = data[['runtime','original_title']]
runtime_order.sort_values(by=['runtime'],ascending = False)
```

> **QUESTION 3b: What is the average runtime of the movies**

```python
#Calculate average runtime for movies
runtime_order['runtime'].mean()
```

```python
#Get information from runtime data.
runtime_order['runtime'].describe()
```

```python
#Plotting runtime of movies and generate the plot with seaborn

plt.figure(figsize=(10,5))
#using seaborn to generate the pl
sns.swarmplot(runtime_order['runtime'], color = 'blue')
plt.show()
```

> Looking at the above calculation, description and plot, we see the plot visually reflect the overall average of 138,25

> 1. 25% of these movies have a runtime of about 120 minutes.
> 2. 50% of these movies have a runtime of about 136 minutes.
> 3. 75% of these movies have a runtime of about 151 minutes.
